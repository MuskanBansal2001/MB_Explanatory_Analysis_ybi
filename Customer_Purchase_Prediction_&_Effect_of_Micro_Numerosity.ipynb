```python
y = purchase_data['Purchased']
```

```python
x = purchase_data[['Age', 'Gender', 'Education', 'Review']]
```

```python
#Encoding categorical variable
one = x.replace({'Review':{'Poor':0, 'Average':1, 'Good':2}}, inplace=True)
x.replace({'Education':{'School':0, 'UG':1, 'PG':2}}, inplace=True)
x.replace({'Gender':{'Male':0, 'Female':1}}, inplace=True)
```

```python
purchase_data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(y,y, train_size=0.7, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
